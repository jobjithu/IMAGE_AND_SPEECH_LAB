```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('imgs/lady.jpg')
```

```python
blur = cv2.blur(img,(5,5))
plt.subplot(121)
plt.title("Original img")
plt.imshow(img)
plt.xticks([]),plt.yticks([])

plt.subplot(122)
plt.title('Blurred img')
plt.imshow(blur)
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
img = cv2.imread('imgs/lady2.jpg')

filter_ = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
sharped_img = cv2.filter2D(img,-1,filter_)

plt.subplot(121)
plt.title("Original img")
plt.imshow(img)
plt.xticks([]),plt.yticks([])

plt.subplot(122)
plt.title('Sharpening img')
plt.imshow(sharped_img)
plt.xticks([]),plt.yticks([])
plt.show()
```
