```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('imgs/baboon.jpg')
```

```python
filter_ = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
```

```python
verti_img = cv2.filter2D(img,-1,filter_)
hori_img = cv2.filter2D(img,-1,np.flip(filter_.T,axis=0))
```

```python
plt.subplot(121)
plt.title('Vertical img')
plt.imshow(verti_img,"gray")
plt.xticks([]),plt.yticks([])

plt.subplot(122)
plt.title('Horizontal img')
plt.imshow(hori_img,"gray")
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
#CANNYMETHOD

img = cv2.imread('imgs/baboon.jpg')
plt.subplot(121)
plt.imshow(img)
plt.xticks([]),plt.yticks([])

plt.subplot(122)
canny = cv2.Canny(img,100,200)
plt.imshow(canny,"gray")
plt.xticks([]),plt.yticks([])
plt.show()
```
