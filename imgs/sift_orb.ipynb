```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("imgs/ship.jpg")
```

```python
plt.subplot(121)
plt.imshow(img)
plt.title("Original img")

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

sift = cv2.SIFT_create()
kp = sift.detect(gray,None)
img1 = cv2.drawKeypoints(gray,kp,img)
plt.subplot(122)
plt.imshow(img1)
plt.show()
```

```python
query_img = cv2.imread("imgs/ship.jpg")
train_img = cv2.imread("imgs/ship.jpg")
```

```python
query_img_gray = cv2.cvtColor(query_img,cv2.COLOR_BGR2GRAY)
train_img_gray = cv2.cvtColor(train_img,cv2.COLOR_BGR2GRAY)
```

```python
orb = cv2.ORB_create()
```

```python
Skeypoints,sdes = orb.detectAndCompute(query_img_gray,None)
Tkeypoints,tdes = orb.detectAndCompute(train_img_gray,None)
```

```python
matcher = cv2.BFMatcher()
matches = matcher.match(sdes,tdes)
```

```python
final_img = cv2.drawMatches(query_img,Skeypoints,train_img,Tkeypoints,matches[:20],None)
final_img = cv2.resize(final_img,(1000,650))
plt.imshow(final_img)
plt.show()
```
