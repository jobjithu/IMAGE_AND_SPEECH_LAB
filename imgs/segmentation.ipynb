```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("imgs/butterfly.jpg")
```

```python
plt.subplot(121)
plt.imshow(img)
plt.title("Original img")
print(img.shape)

pixel_val = img.reshape((-1,3))
pixel_val = np.float32(pixel_val)

criteria = (cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER,100,0.85)
k = 3
retval,labels,centers = cv2.kmeans(pixel_val,k,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]
segmented_img = segmented_data.reshape(img.shape)
print(segmented_img.shape)

plt.subplot(122)
plt.imshow(segmented_img)
plt.title("Clustered img")
plt.show()
```
