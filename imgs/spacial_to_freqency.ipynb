```python
#fourier transform
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('imgs/ship.jpg',0)
img2 = np.fft.fft2(img1)
img3 = np.fft.fftshift(img2)
img4 = np.fft.ifftshift(img3)
img5 = np.fft.ifft2(img4)
plt.figure(figsize=(6*4,5*4))
plt.subplot(151),plt.title("Original img")
plt.imshow(img1,"gray")

plt.subplot(152),plt.title("Spectrum")
plt.imshow(np.log(1+np.abs(img2)),"gray")

plt.subplot(153),plt.title("Centered spectrum")
plt.imshow(np.log(1+np.abs(img3)),"gray")

plt.subplot(154),plt.title("Decenterlized")
plt.imshow(np.log(1+np.abs(img4)),"gray")

plt.subplot(155),plt.title("Process spectrum")
plt.imshow(np.abs(img5),'gray')

plt.show()
```
