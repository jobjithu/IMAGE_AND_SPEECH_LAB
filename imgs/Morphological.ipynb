```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("imgs/aa.tiff")
```

```python
kernel = np.ones((2,3),np.uint8)
```

```python
erosion = cv2.erode(img,kernel,iterations = 1)
dilation = cv2.dilate(img,kernel,iterations = 1)
```

```python
plt.subplot(131)
plt.imshow(img)
plt.title("Original img")

plt.subplot(132)
plt.imshow(erosion)
plt.title("Erotedimg")

plt.subplot(133)
plt.imshow(dilation)
plt.title("dilated img")
plt.show()
```
