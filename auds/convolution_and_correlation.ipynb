```python
#convolution

import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
d_file = 'set1.wav'
d1_file = 'set2.wav'
```

```python
d,sr = librosa.load(d_file,sr =8000)
t = np.linspace(0,len(d)/sr,len(d))
plt.subplot(221)
plt.plot(t,d,lw =1)
plt.xlabel('Time')
plt.ylabel("Amplitude")
plt.title("file 1")
plt.grid(True)


d1,sr = librosa.load(d1_file,sr =8000)
t1 = np.linspace(0,len(d1)/sr,len(d1))
plt.subplot(222)
plt.plot(t1,d1,lw =1)
plt.xlabel('Time')
plt.title('file 2')
plt.ylabel("Amplitude")
plt.grid(True)

y = np.convolve(d,d1)
ty = np.linspace(0,len(y)/sr,len(y))
plt.subplot(223)
plt.plot(ty,y,lw =1)
plt.title('Convolution')
plt.xlabel('Time')
plt.ylabel("Amplitude")
plt.grid(True)

plt.show()

```

```python
#auto correlation

d_file = 'speech.wav'

d,sr = librosa.load(d_file,sr =8000)
t = np.linspace(0,len(d)/sr,len(d))
plt.subplot(211)
plt.plot(t,d,lw =1)
plt.xlabel('Time')
plt.ylabel("Amplitude")
plt.title("file 1")
plt.grid(True)


r = librosa.autocorrelate(d,max_size=1000)
plt.subplot(212)
plt.acorr(r,maxlags = 20)
plt.xlabel("lags")
plt.ylabel("Coefficients")
plt.title("auto correlate")
plt.grid(True)
plt.show()
```

```python
#crosscorrelate

d_file = 'sample1.wav'
d1_file = 'sample2.wav'

d,sr = librosa.load(d_file,sr =8000)
t = np.linspace(0,len(d)/sr,len(d))
plt.subplot(221)
plt.plot(t,d,lw =1)
plt.xlabel('Time')
plt.ylabel("Amplitude")
plt.title("file 1")
plt.grid(True)


d1,sr = librosa.load(d1_file,sr =8000)
t1 = np.linspace(0,len(d1)/sr,len(d1))
plt.subplot(222)
plt.plot(t1,d1,lw =1)
plt.xlabel('Time')
plt.title('file 2')
plt.ylabel("Amplitude")
plt.grid(True)

correlation = np.correlate(d,d1,mode = "full")
lags = np.arange(-100,100)
plt.subplot(223)
plt.plot(lags,correlation[0:200])
plt.xlabel("Lags")
plt.ylabel("Coefficients")
plt.grid(True)
plt.show()
```
