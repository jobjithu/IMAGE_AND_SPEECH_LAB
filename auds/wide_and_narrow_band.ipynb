```python
import librosa.core as lc

import matplotlib.pyplot as plt
import librosa
import numpy as np
```

```python
d_file = 'speech.wav'
y,sr = librosa.load(d_file,sr = 8000)
t = np.linspace(0,len(y)/sr,len(y))
plt.subplot(211)
plt.plot(t,y,lw =1)
plt.xlabel("Time")
plt.ylabel("Amplitude")

n_fft  = 1024
fig,ax = plt.subplots()

mag = np.abs(lc.stft(y,n_fft = n_fft,hop_length = 10,win_length = 40,window = 'hamming'))
mag1 = librosa.amplitude_to_db(mag,ref = np.max)
img = librosa.display.specshow(mag1,sr = sr ,x_axis='s',y_axis='linear')
ax.set(title = 'wideband')
fig.colorbar(img,ax = ax , format = "% 2.f dB")

fig,ax = plt.subplots()

mag1 = np.abs(lc.stft(y,n_fft = n_fft,hop_length = 100,win_length = 400,window = 'hamming'))
D1 = librosa.amplitude_to_db(mag1,ref = np.max)
img = librosa.display.specshow(D1,sr = sr ,x_axis='s',y_axis='linear')
ax.set(title = 'narrowband')
fig.colorbar(img,ax = ax , format = "% 2.f dB")
plt.show()
```
