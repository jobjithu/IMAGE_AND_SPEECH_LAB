```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
from scipy import signal as sg
```

```python
d_file = 'set1.wav'
```

```python
d,sr = librosa.load(d_file,sr = 8000)
t = np.linspace(0,len(d)/sr,len(d))
plt.subplot(211)
plt.plot(t,d,lw =1)
plt.title('Original file')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)

fig,ax = plt.subplots(1,1,figsize=(6,3))
b,a = sg.butter(4,2000./(sr/2.),'high')
d_fil = sg.filtfilt(b,a,d)
t = np.linspace(0.,len(d)/sr,len(d))
ax.plot(t,d,lw =1)
ax.plot(t,d_fil,lw = 1)
plt.title('Hight pass')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)

```
